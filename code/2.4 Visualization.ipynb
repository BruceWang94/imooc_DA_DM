## Import package and load data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/HR.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Bar Chart 

```python
plt.bar(np.arange(len(df['salary'].value_counts())), df['salary'].value_counts())
plt.show()
```

```python
plt.title('SALARY')
plt.xlabel('salary')
plt.ylabel('Number')
plt.xticks(np.arange(len(df['salary'].value_counts())), df['salary'].value_counts().index)
plt.bar(np.arange(len(df['salary'].value_counts())), df['salary'].value_counts())
plt.show()
```

```python
plt.title('SALARY')
plt.xlabel('salary')
plt.ylabel('Number')
plt.xticks(np.arange(len(df['salary'].value_counts()))+0.5, df['salary'].value_counts().index)
plt.axis([0,4,0,10000])
plt.bar(np.arange(len(df['salary'].value_counts()))+0.5, df['salary'].value_counts(), width=0.5)
plt.show()
```

```python
plt.title('SALARY')
plt.xlabel('salary')
plt.ylabel('Number')
plt.xticks(np.arange(len(df['salary'].value_counts()))+0.5, df['salary'].value_counts().index)
plt.axis([0,4,0,10000])
plt.bar(np.arange(len(df['salary'].value_counts()))+0.5, df['salary'].value_counts(), width=0.5)
for x,y in zip(np.arange(len(df['salary'].value_counts()))+0.5, df['salary'].value_counts()):
    plt.text(x,y,y,ha='center', va='bottom')
plt.show()
```

### Use seaborn to plot the graph

```python
sns.set()
sns.set_style(style='whitegrid')
sns.set_context(context='poster',font_scale=0.8)
sns.set_palette(sns.color_palette('RdBu',n_colors=7))
sns.countplot(x='salary', data=df)
plt.show()
```

```python
sns.set()
sns.set_style(style='whitegrid')
sns.set_context(context='poster', font_scale=0.8)
sns.set_palette(sns.color_palette('RdBu', n_colors=7))
sns.countplot(x='salary', hue='department', data=df)
plt.show()
```

## Histgram

```python
## move the outlier
df = df.dropna(axis=0, how='any')
df = df[df['last_evaluation']<=1][df['salary']!='nme'][df['department']!='sale']
```

```python
sns.set()
f = plt.figure()
f.add_subplot(1,3,1)
sns.distplot(df['satisfaction_level'], bins=10) # two more parameters hist=False or kde=False
plt.show()
```

```python
sns.set()

f = plt.figure()
f.add_subplot(1,3,1)
sns.distplot(df['satisfaction_level'], bins=10) # two more parameters hist=False or kde=False
f.add_subplot(1,3,2)
sns.distplot(df['last_evaluation'], bins=10)
f.add_subplot(1,3,3)
sns.distplot(df['average_monthly_hours'], bins=10)
plt.subplots_adjust(left=None, bottom=None, right=2, top=None, wspace=None, hspace=None)
plt.show()
```

## Boxplot

```python
sns.set()

sns.boxplot(y=df['time_spend_company'])
plt.show()
```

```python
sns.set()

sns.boxplot(x=df['time_spend_company'], saturation=0.75, whis=3) # the parameter 'whis' is the k in formula
plt.show()
```

## Line Chart

```python
sns.set()

sub_df = df.groupby('time_spend_company').mean()
sns.pointplot(sub_df.index, sub_df['left'])
plt.show()
```

```python
sns.set()

sns.pointplot(x='time_spend_company', y='left', data=df)
plt.show()
```

## Pie Chart

```python
sns.set()

plt.pie(df['department'].value_counts(normalize=True))
plt.show()
```

```python
## add label
sns.set()
lbs = df['department'].value_counts().index
plt.pie(df['department'].value_counts(normalize=True), labels=lbs)
plt.show()
```

```python
## add value
sns.set()
lbs = df['department'].value_counts().index
plt.pie(df['department'].value_counts(normalize=True), labels=lbs, autopct='%1.1f')
plt.show()
```

```python
## change color
sns.set()
lbs = df['department'].value_counts().index
plt.pie(df['department'].value_counts(normalize=True), 
        labels=lbs, 
        autopct='%1.1f',
        colors=sns.color_palette('Reds'))
plt.show()
```

```python
## seperate sale
sns.set()
lbs = df['department'].value_counts().index
explodes=[0.1 if i=='sales' else 0 for i in lbs]
plt.pie(df['department'].value_counts(normalize=True), explode=explodes, labels=lbs, autopct='%1.1f',colors=sns.color_palette('Reds'))
plt.show()
```

```python
## Or check 'salary'
sns.set()
lbs = df['salary'].value_counts().index
explodes=[0.1 if i=='low' else 0 for i in lbs]
plt.pie(df['salary'].value_counts(normalize=True), explode=explodes, labels=lbs, autopct='%1.1f',colors=sns.color_palette('Reds'))
plt.show()
```
