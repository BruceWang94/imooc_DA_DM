## Import package

```python
import pandas as pd
import numpy as np
```

## Load data

```python
## load data
df = pd.read_csv('./data/HR.csv')
```

```python
df
```

## Drop the outlier

First, to handle the null value. We use `dropna()`, there are two parameters, `axis` and `how`:

`axis`:
- 0 - drop the row
- 1 - drop the column

`how`:
- 'all' - if all the cell are null, then drop
- 'any' - if there exist one null, then drop 

```python
df = df.dropna(axis=0, how='any')
```

```python
df
```

```python
df = df[df['last_evaluation']<=1][df['salary']!='nme'][df['department']!='sale']
```

## Sample Comparative Analysis

The comparative analysis is based on the department, we use `groupby()` function to group by the data. After the function, it needs an aggregation method.

```python
df.groupby('department').mean()
```

```python
df.loc[:, ['last_evaluation','department']].groupby('department')
```

```python
df.loc[:, ['last_evaluation','department']].groupby('department').mean()
```

We can also define the aggregation function by ourselves.

```python
df.loc[:, ['average_monthly_hours', 'department']].groupby('department')['average_monthly_hours'].apply(lambda x: x.max() - x.min())
```
