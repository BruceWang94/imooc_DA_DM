```python
## use quantile to filter the outliers
le_s = df['last_evaluation']

## cal quantile
q_low = le_s.quantile(q=0.25)
q_high = le_s.quantile(q=0.75)

q_interval = q_high - q_low
k = 1.5
```

```python
le_s = le_s[le_s<q_high+k*q_interval][le_s>q_low-k*q_interval]
```
